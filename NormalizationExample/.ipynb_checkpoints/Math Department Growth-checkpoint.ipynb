```python
#show plot in the ipython notebook
%matplotlib inline

#import basic libraries
import numpy as np
import matplotlib.pyplot as mpl
from pandas import read_excel, DataFrame

#import the data
df = read_excel("Math Department Data.xlsx")
df
```

```python
#get the exact column you want to visualize, in this case "Math/Econ"
dfME = df["Math/Econ"]

#plot Math Econ
fME = mpl.plot(dfME)

#add custom axis ranges
f = mpl.xticks(np.arange(14) + np.array(np.ones(14)*.5), df["Quarter"])
f = mpl.yticks(np.arange(0,280,20))


#add titles
f = mpl.title("Growth of Math/Economics Majors")
f = mpl.xlabel("Quarter")
f = mpl.ylabel("Number of Students")

#unnecessary if using "%matplotlib inline"
#mpl.show()
```

```python
dfT = df["TOTAL"]

#plot Math Econ and the Math department as a whole for perspective
fME = mpl.plot(dfME)
fT = mpl.plot(dfT, color = 'red')

#add custom axis ranges
f = mpl.xticks(np.arange(14) + np.array(np.ones(14)*.5), df["Quarter"])

#add titles
f = mpl.title("Growth (?) of Math/Economics Majors")
f = mpl.xlabel("Quarter")
f = mpl.ylabel("Number of Students")
f = mpl.legend( (fME[0], fT[0]), ("Math/Economics", "Math Department"))

#unnecessary if using "%matplotlib inline"
#mpl.show()
```

```python
#percentages are the simplest way to normalize data
#initialize array
dfP = np.zeros(15)

#populate array
for x in np.arange(0,15,1):
    dfP[x] = dfME[x] / dfT[x]

#plot it!
f3 = mpl.plot(dfP)

#add custom axis ranges
f3 = mpl.yticks( np.arange(0,1.1,.1))
f3 = mpl.xticks( np.arange(14) + np.array(np.ones(14)*.5), df["Quarter"])

#add titles
f = mpl.title("Growth (?) of Math/Economics Majors")
f = mpl.xlabel("Quarter")
f = mpl.ylabel("Percentage of Students within Math Department")

#unnecessary if using "%matplotlib inline"
#mpl.show()
```
